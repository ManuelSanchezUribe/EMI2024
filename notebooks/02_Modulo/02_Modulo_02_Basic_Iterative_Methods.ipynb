## Objective: Review basic iterative Method for solving the linear systems.

Let $A\in \mathbb R^{n\times n}$ and $b\in \mathbb R^{n}$. Consider the problem of finding $x\in \mathbb R^{n}$ such that

$$
Ax = b
$$



### General splitting for iteration methods.

A basic iterative method given by the iteration
$$
M x_{k+1} = N x_{k} + b
$$
can be defined for any splitting of the form $A = M- N$ where $M$ is nonsingular. The matrix $G = M^{-1}N$ is known as the *iteration matrix* of the methods,
$$
x_{k+1} = G x_{k} + f, \quad f = M^{-1} b
$$

If the above iteration converges, its limit $x$ satisfies

$$
x = Gx + f
$$



### General splitting for iteration methods.

In the case where the above iteration arises from the splitting $A=M-N$, it is easy to see that the solution $x$ to the above system is identical to that of the original system $Ax=b$. Indeed, in this case the sequence has the from

$$
x_{k+1} = M^{-1} N x_{k} + M^{-1} b
$$

and its limit satisfies 

$$
Mx = Nx +b \quad \Longleftrightarrow \quad Ax = b
$$



### General convergence result

***Theorem***. Let $G$ be a square matrix such that $\rho(G)<1$. Then $(I-G)$ is nonsingular and the iteration $x_{k+1} = G x_{k} + f$ converges for any $f$ and $x_0$. Conversely, if the iteration converges for any $f$ and $x_0$, then $\rho(G)<1$.

Here the spectal radius $\rho(A)$ of a matrix $A$ is defined as the maximum modulus of the eigenvalues of $A$, i.e.

$$
\rho(A) = \max_{\lambda \in \sigma(A)} |\lambda|
$$

### General convergence result

***Corollary***. Let $G$ be a square matrix such that $\|G\|<1$ for some matrix norm $\|\cdot\|$. Then $(I-G)$ is nonsingular and the iteration converges for any initial vector $x_0$.

$$
d_{k} = x_{k} - x_{*} = G x_{k-1} + f - G x_{*} - f = G d_{k-1} = \ldots = G^{k} d_0

A general convergence factor:

$$
\phi = \lim_{k\to \infty} \left( \max_{x_0\in \mathbb R^{n} } \frac{\|d_k\|}{\|d_0\|}\right)^{1/k}
$$

In this case

$$
\phi = \lim_{k \to \infty}\left( \max_{x_0\in \mathbb R^{n} } \frac{\|G^{k} d_0\|}{\|d_0\|}\right)^{1/k} = \lim_{k \to \infty}\left( \|G^{k}\|\right)^{1/k} = \rho(G).
$$

The global asymptotic factor is equal to the spectral radius of the iteration matrix.

#### Example: Richardson's iteration
For $\alpha \geq 0$, consider the iteration

$$
x_{k+1} = x_{k} +\alpha (b-A x_{k})\quad \Longrightarrow \quad x_{k+1} = \underbrace{(I-\alpha A)}_{G_{\alpha}}x_{k} + \alpha b
$$

The convergence factor is $\rho(I-\alpha A)$. Assume that the eigenvalues $\lambda_i$, $1\leq i\leq n$,  of $A$ are all real and such that

$$\lambda_{\min} \leq \lambda_i \leq \lambda_{\max} $$

Then, the eigenvalues $\mu_i$ of $G_{\alpha}$ are such that

$$ 1- \alpha\lambda_{\max} \leq \mu_i \leq 1 - \alpha \lambda_{\min} $$

#### Example: Richardson's iteration 

If $\lambda_{\min} < 0$ and $\lambda_{\max} >0 $ , at least one eigenvalue is $>1$, and so $\rho(G_{\alpha})>1$, for any $\alpha$,and the method diverges for any $x_0$.

If all the eigenvalues are positive, $\lambda_{\min}>0$, then the method converges for any scalar $\alpha$ which satisfies

$$
0<\alpha<2/\lambda_{\max}.
$$

Optimal $\alpha$?

$$
\alpha_{\text{opt}} = \frac{2}{\lambda_{\min} + \lambda_{\max}}
$$

```python
from ngsolve import *
from ngsolve.webgui import Draw
from ngsolve.la import EigenValues_Preconditioner
import matplotlib.pyplot as plt
```

```python
# Poisson problem
def Setup(h=0.01, p=1):
    mesh = Mesh(unit_square.GenerateMesh(maxh=h))
    fes = H1(mesh, order=p, dirichlet="left|bottom")
    u, v = fes.TnT()
    a = BilinearForm(grad(u)*grad(v)*dx + u*v*dx)
    f = LinearForm(v*dx)
    gfu = GridFunction(fes)
    return mesh, fes, a, f, gfu
```

```python
def RichardsonIteration(gfu, a, b, alpha, maxits=1000, tol=1e-8):
    fes = gfu.space
    du = gfu.vec.CreateVector()
    res = gfu.vec.CreateVector()

    for it in range(maxits):
        res[:] = 0
        res.data = -a.mat*gfu.vec + b.vec
        du.data = alpha*res
        gfu.vec.data += du
        
        #stopping criteria
        stopcritval = sqrt(abs(InnerProduct(du,du)))
        print ("<du",it,", du",it,">_{-1}^0.5 = ", stopcritval)
        if stopcritval < tol and it > 1:
            break
    return it+1
```

```python
def JacobiIteration(gfu, a, b, maxits=1000, tol=1e-8):
    fes = gfu.space
    du = gfu.vec.CreateVector()
    res = gfu.vec.CreateVector()
    diag = (a.mat.CreateSmoother(fes.FreeDofs()))
    for it in range(maxits):
        res[:] = 0
        res.data = -a.mat*gfu.vec + b.vec
        du.data = diag*res
        gfu.vec.data += du
        
        #stopping criteria
        stopcritval = sqrt(abs(InnerProduct(du,du)))
#         stopcritval = sqrt(abs(InnerProduct(res,res)))
        print ("<du",it,", du",it,">_{-1}^0.5 = ", stopcritval)
        if stopcritval < tol and it > 1:
            break
    return it+1
```

```python
def GaussSeidelIteration(gfu, a, b, maxits=1000, tol=1e-8):
    gfu.vec[:] = 0
    fes = gfu.space
    res = b.vec.CreateVector()              # residual
    preJpoint = a.mat.CreateSmoother(fes.FreeDofs())
    projres = b.vec.CreateVector()          # residual projected to freedofs
    proj = Projector(fes.FreeDofs(), True)
    
    for it in range(maxits):
        preJpoint.Smooth(gfu.vec, b.vec)    # one step of point Gauss-Seidel
        res.data = b.vec - a.mat*gfu.vec
        projres.data = proj * res

        #stopping criteria
        stopcritval =Norm(projres)# sqrt(abs(InnerProduct(du,du)))
#         stopcritval = sqrt(abs(InnerProduct(res,res)))
        print ("<du",it,", du",it,">_{-1}^0.5 = ", stopcritval)
        if stopcritval < tol and it > 1:
            break
    return it+1
```

```python
mesh, fes, a, b, gfu = Setup(h=0.05, p=1)
a.Assemble()
b.Assemble()

Draw(gfu)
```

```python
gauss_seidel_its = GaussSeidelIteration(gfu, a, b, maxits=5000)
```

```python
preI = Projector(mask=fes.FreeDofs(), range=True)
lams = EigenValues_Preconditioner(mat=a.mat, pre=preI)
print(f"lambda max: {max(lams)}, lambda min: {min(lams)}" )
alphaopt = 2/(max(lams)+min(lams))
richardson_its = RichardsonIteration(gfu, a, b, alpha=alphaopt, maxits=5000)
```

```python
jacobi_its = JacobiIteration(gfu, a, b, maxits=5000)
```

```python
print("Richardson iterations:", richardson_its)
print("Jacobi iterations:", jacobi_its)
print("Gauss-Seidel iterations:", gauss_seidel_its)
```
