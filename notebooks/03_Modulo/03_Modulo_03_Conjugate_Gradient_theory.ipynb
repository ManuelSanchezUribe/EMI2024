## Objective: Iterative algorithm for solving sparse symmetric and positive definite linear systems

Orthogonal projection technique onto the Krylov subspace $\mathcal K_{m}(r_0, A)$ , $r_0  = b - A x_0$. 

Let $A\in \mathbb R^{n\times n}$ be a symmetric positive definite matrix.

We begin introducing a simple algorithm, gradient descent.

### Steepest descent algorithm:
\begin{align*}
& r \gets b-Ax \\
& \alpha \gets \|r\|^{2}_2/ (Ar\cdot r) \\
& x\gets x+\alpha r
\end{align*}

```python
def SteepestDescent(A, b, x0, TOL=1e-8, MAXIT=100, print_iter=False):
    x = x0.copy()
    g = A.dot(x0) - b 
    ng = g.dot(g)
    Xlist = []
    Xlist.append(x.copy())
    niter = 0
    while np.sqrt(ng)>TOL and niter < MAXIT:
        # step length corresponding to exact line search
        mu = ng / np.dot(g, np.dot(A,g))
        x -= mu*g
        g = A.dot(x) - b
        ng = g.dot(g)
        niter += 1
        if print_iter: printSD(niter,ng)
        Xlist.append(x.copy())
    return x, niter, Xlist
def printSD(iter, res):
    print(f" *** SD iteration {iter}, residual = {res:1.2e}")
```

### Ejemplo 1:

Find solution $x$ such that $Ax = b$, for $A$ symmetric positive definite. This is equivalent to minimize the quadratic function $\phi(x) = \frac{1}{2} x^\top A x - x^\top b$.

In particular the case
\begin{equation*} A = \left(\begin{array}{cc} a_{11}&0\\ 0 &a_{22}\end{array}\right), \quad b = \left( \begin{array}{c} 0\\0 \end{array}\right) \end{equation*}

We need to minimize the function
\begin{equation*}
\phi(x) = \frac{1}{2} x^\top A x - x^\top b = \frac{1}{2}(a_{11} x^2 + a_{22} y^2)
\end{equation*}

```python
# Example 1: Data 
a11 = 1.0; a22  = 0.1
A = np.array([[a11,.0],[.0,a22]]); b = np.array([0.0,.0])

# Iteracion Steepest descent
x, n, Xlist = SteepestDescent(A,b,x0=np.array([a22,1]), print_iter=False)
```

```python
# function
phi = lambda x: 0.5*x.dot(A.dot(x)) - x.dot(b)

# Make data.
x = np.arange(-1.2, 1.2, 0.025)
y = np.arange(-1.2, 1.2, 0.025)
X, Y = np.meshgrid(x, y)

Z = 0.5*(a11*X**2 + a22*Y**2)
fig, ax = plt.subplots(1,1, figsize=(10,8))

#  Steepest descent points
X0 = Xlist[0]; x0 = X0[0]; y0 = X0[1]; f0 = phi(Xlist[0])
X1 = Xlist[1]; x1 = X1[0]; y1 = X1[1]; f1 = phi(Xlist[1])
X2 = Xlist[2]; x2 = X2[0]; y2 = X2[1]; f2 = phi(Xlist[2])
X3 = Xlist[3]; x3 = X3[0]; y3 = X3[1]; f3 = phi(Xlist[3])
X4 = Xlist[4]; x4 = X4[0]; y4 = X4[1]; f4 = phi(Xlist[4])
X5 = Xlist[5]; x5 = X5[0]; y5 = X5[1]; f5 = phi(Xlist[5])
X6 = Xlist[6]; x6 = X6[0]; y6 = X6[1]; f6 = phi(Xlist[6])
X7 = Xlist[7]; x7 = X7[0]; y7 = X7[1]; f7 = phi(Xlist[7])
CS = ax.contour(X,Y, Z, levels=[f7,f6,f5,f4,f3,f2,f1,f0], cmap='copper')
ax.clabel(CS, fontsize=9, inline=True)
ax.plot([x0,x1],[y0,y1],'--*',color='C00', linewidth=3)
ax.plot([x1,x2],[y1,y2],'--*',color='C00', linewidth=3)
ax.plot([x2,x3],[y2,y3],'--*',color='C00', linewidth=3)
ax.plot([x3,x4],[y3,y4],'--*',color='C00', linewidth=3)
ax.plot([x4,x5],[y4,y5],'--*',color='C00', linewidth=3)
ax.plot([x5,x6],[y5,y6],'--*',color='C00', linewidth=3)
ax.plot([x6,x7],[y6,y7],'--*',color='C00', linewidth=3, label='steepest descent')

ax.set_aspect('equal')
ax.legend()
plt.savefig('fig1steepest_descent.png')
plt.show()
```

### Conjugate Gradient Method

\begin{align*}
& \alpha_j \gets \|r_j\|_2^2/(Ap_j \cdot p_j)\\
& x_{j+1} \gets x_{j} + \alpha_j p_j \\
& r_{j+1} \gets r_{j} - \alpha_j A p_j \\
& \beta_{j}\gets\|r_{j+1}\|_2^2/\|r{j}\|_2^2\\
& p_{j+1} \gets r_{j+1} + \beta_{j}p_{j}
\end{align*}

```python
def ConjugateGradient(A, b, x0 = None, TOL=1e-8, MAXIT = 100, print_iter=True):
    '''
    Algorithm 11.3.3. Conjugate Gradients: Hestenes-Stiefel Version
    Input  : A s.d.p, b, x0
    Output : x sol. Ax = b
    '''
    if x0 is None: x0 = np.zeros(b.size)
    k = 0
    r0 = b - A.dot(x0) 
    # vectors
    x = []; r = []
    x.append(x0); r.append(r0); pk=r0.copy()
    for k in range(MAXIT):
        alphak =  r[k].dot(r[k])/(pk.dot(A.dot(pk)))
        xnew = x[k] + alphak * pk; x.append(xnew)
        rnew = r[k] - alphak*A.dot(pk); r.append(rnew)
        betak = rnew.dot(rnew)/(r[k].dot(r[k]))
        pk = rnew + betak*pk
        if print_iter: printCG(k, np.linalg.norm(rnew))
        if np.linalg.norm(rnew)<TOL:
            break
    return x, k
def printCG(iter, res):
    print(f" *** CG iteration {iter}, residual = {res:1.2e}")   
```

```python
# Example 1: Data 
a11 = 1.0; a22  = 0.1
A = np.array([[a11,.0],[.0,a22]]); b = np.array([0.0,.0])

# Steepest descent Iteration
x, n, Xlist = SteepestDescent(A,b,x0=np.array([a22,1]), print_iter=False)

# CG Iteration
xcg, niter = ConjugateGradient(A, b, x0=np.array([a22,1.0]), print_iter=True)
print("residual \|Ax-b\|:", np.linalg.norm(A.dot(xcg[-1])-b))
```

## Two examples, bounded vs. increasing condition number

Consider the following matriz in terms of the coefficient $a,b,c$ and the size $m$
$$
A = I_{m} \otimes C_1 + C_2 \otimes I_{m},
$$ 

$$ C_{1}
=
\begin{bmatrix}
c      &     a  &      0 & \ldots & 0     \\
a      &     c  &      a & \ddots & \vdots\\
0      & \ddots & \ddots & \ddots & 0     \\
\vdots & \ddots & a      & c      & a     \\
0      & \ldots & 0      & a      & c
\end{bmatrix}, \quad
C_{2}
=
\begin{bmatrix}
c      &     b  &      0 & \ldots & 0     \\
b      &     c  &      b & \ddots & \vdots\\
0      & \ddots & \ddots & \ddots & 0     \\
\vdots & \ddots & b      & c      & b     \\
0      & \ldots & 0      & b      & c
\end{bmatrix}
$$

1. Test with bounded condition number ($\kappa(A)<9$): $a= 1/9,\,b = 1/0,\, c = 5/18$.

2. Test with increasing condition number : $a= -1,\,b = -1,\, c = 2$.


```python
import pandas as pd
def testproblem(num, m):
    if num == 2: # Poisson matrix
        a = -1
        b = a
        c = 2
    elif num == 1:# Averaging problem
        a = 1.0/9.0
        b = a
        c = 5.0/18.0
    C1 = c*np.diag(np.ones(m),0)+a*np.diag(np.ones(m-1),-1)+a*np.diag(np.ones(m-1),1);
    C2 = c*np.diag(np.ones(m),0)+b*np.diag(np.ones(m-1),-1)+b*np.diag(np.ones(m-1),1)
    A = np.kron(np.eye(m),C1) + np.kron(C2, np.eye(m))
    bvector = np.random.rand(m*m)
    return sp.csr_matrix(A), bvector
```

```python
A,_=testproblem(num=2, m=40)
plt.spy(A)
```

```python
def TestCG_example(examplenumber, mm=5):
    kappa = np.zeros(mm)
    nn =np.zeros(mm, dtype=np.int64)
    M = [5*2**i for i in range(mm)]
    for m in range(len(M)):
        A, b = testproblem(examplenumber, M[m])
        x, nn[m] = ConjugateGradient(A,b, MAXIT=1000, print_iter=False)
        kappa[m] = np.linalg.cond(A.todense())
    tab = pd.DataFrame({'m':M, 'Matrix dim':np.square(M), 'Condition number':kappa, 'number of CG iterations': nn})
    return tab
```

### Using scipy.sparse.linlg.cg

```python
class cg_counter(object):
    def __init__(self, A, b, disp=True):
        self._disp = disp
        self.niter = 0
        self.A = A
        self.b = b
        self.rk = []
    def __call__(self, xk=None):
        self.niter += 1
        if self._disp:
            rk = self.b-self.A.dot(xk) 
            rhok = np.linalg.norm(rk)
            # print('iter %3i\trk = %s' % (self.niter, str(rhok)))      
            self.rk.append(rk)
```

```python
A, b = testproblem(1, 80)
counter = cg_counter(A, b)
x, info = cg(A, b, tol=10**(-8), maxiter=200, callback=counter)
counter.niter
```

```python
def Test_scipyCG_example(examplenumber, mm=5):
    kappa = np.zeros(mm)
    nn =np.zeros(mm, dtype=np.int64)
    M = [5*2**i for i in range(mm)]
    for m in range(len(M)):
        A, b = testproblem(examplenumber, M[m])
        counter = cg_counter(A, b)
        x, info = cg(A,b, tol=10**(-8), maxiter=200, callback=counter)
        nn[m] = counter.niter
        kappa[m] = np.linalg.cond(A.todense())
    tab = pd.DataFrame({'m':M, 'Matrix dim':np.square(M), 'Condition number':kappa, 'number of CG iterations': nn})
    return tab
```
